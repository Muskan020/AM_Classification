```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
# Replace 'your_file.csv' with the actual path to your CSV file
data = pd.read_csv('/kaggle/input/pred-trueam/true_and_predicted_labels.csv')

# Assuming 'True Label' column has the true labels and 'Predicted Label' has the predicted labels (multiclass)
y_true = data['True Label']  # Replace with your actual column name for true labels
y_pred_proba = data['Predicted Label']  # Replace with your predicted probabilities column

# Binarize the true labels for multiclass (One-vs-Rest approach)
classes = sorted(y_true.unique())  # Get unique classes
y_true_binarized = label_binarize(y_true, classes=classes)

# Calculate ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i, class_label in enumerate(classes):
    fpr[class_label], tpr[class_label], _ = roc_curve(y_true_binarized[:, i], y_pred_proba == class_label)
    roc_auc[class_label] = auc(fpr[class_label], tpr[class_label])

# Plotting the ROC curve for each class
plt.figure()
for class_label in classes:
    plt.plot(fpr[class_label], tpr[class_label],
             lw=2, label=f'ROC curve of class {class_label} (AUC = {roc_auc[class_label]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for Multiclass')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the CSV file
data = pd.read_csv('/kaggle/input/pred-trueam/true_and_predicted_labels.csv')

# Assuming 'True Label' column has the true labels and 'Predicted Label' has the predicted probabilities
y_true = data['True Label']  # Replace with your actual column name for true labels
y_pred_proba = data.iloc[:, 1:].values  # Assuming the remaining columns are predicted probabilities for each class

# Binarize the true labels for multiclass (One-vs-Rest approach)
classes = sorted(y_true.unique())  # Get unique classes
y_true_binarized = label_binarize(y_true, classes=classes)

# Compute ROC curve and ROC area for micro-average
fpr, tpr, _ = roc_curve(y_true_binarized.ravel(), y_pred_proba.ravel())
roc_auc = auc(fpr, tpr)

# Plotting the micro-average ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Micro-average ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Model Level')
plt.legend(loc="lower right")
plt.show()
```
